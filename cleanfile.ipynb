```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv(r"C:\Users\ranji\Downloads\my_file (1).csv")
```

```python
df.columns=df.columns.str.replace('\xa0','',regex=False)#clean the columnname
print([{col} for col in df.columns])
```

```python
df
```

```python
df['Ref.']=df['Ref.'].apply(lambda x: re.sub(r"[^a-zA-Z0-9]",'',x))# substitute any character in the string x that is not [a-zA-Z0-9](removing it)
```

```python
df
```

```python
df['Ref.']=df['Ref.'].str.replace('1516','16')
df['Ref.']=df['Ref.'].str.replace('d','5')
```

```python
df[['Actualgross','Adjustedgross (in 2022 dollars)','Average gross']]
```

```python
def clean_gross(text):
    text=re.sub(r"\[\s*.*?\s*\]","",text)
    text=re.sub(r"[^a-zA-Z0-9]","",text)
    return text
cols=['Actualgross','Adjustedgross (in 2022 dollars)','Average gross']#define a fuction to clean 3 columns in once
for col in cols:
    df[col]=df[col].apply(clean_gross)
```

```python
df
```

```python
df['Tour title']
```

```python
def clean_title(text):
    text=re.sub(r"[^a-zA-Z ]","",text)
    return text
df['Tour title']=df['Tour title'].apply(clean_title)#clean the tour title field has unwanted data
df
```

```python
df['Tour title']
```

```python
df.info()
```

```python
df=df.drop(['Peak','All Time Peak'],axis=1)#drop unwanted columns
```

```python
num_col=['Actualgross','Adjustedgross (in 2022 dollars)','Shows','Average gross']
for co in num_col:
    df[co]=df[co].astype("float")#change the type
```
